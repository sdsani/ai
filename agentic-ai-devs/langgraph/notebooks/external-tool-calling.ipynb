```python
# Sometimes we need a model to interact directly with external systems (examples: Databases/apis)
# These systems require input in a specific format. Next we will explore how to interact with these systems
# This is where tools calling comes in. Tools are just functions

import os, getpass
import logging

def _set_env(var: str):
    logging.warning("Calling setEnv")
    if not os.environ.get(var):
        logging.warning("API Key not found. Let's input")
        os.environ[var] = getpass.getpass(f"{var}: ")
    else:
        logging.warning("Found API Key in the environment")

_set_env("OPENAI_API_KEY")

from langchain_openai import ChatOpenAI

llm_with_no_tools = ChatOpenAI(model="gpt-4.0")

print("Done with creation of an LLM")
```

```python
# Let's create some tools now
import datetime

def shoopeninghourshardcoded() -> str:
    "This tool provide the daily opening hours schedule for a shop."
    schedule = """Monday 10:00am to 08:00pm
                  Tuesday 10:00am to 08:00pm
                  Wednesday 10:00am to 08:00pm
                  Thursday 10:00am to 08:00pm
                  Friday 10:00am to 01:00pm
                  Saturday 11:00am to 05:00pm
                  Sunday 01:00pm to 06:00pm
                """
    
    return schedule;

def Currenttime() -> datetime:
    "This tool provide the current date and time."
    return datetime.datetime.now()

# Wrap functions into a list, making them available for the model.
tools = [shoopeninghourshardcoded,Currenttime]

# Next we have to make sure that model knows about these tools/functions
llm_with_tools = llm_with_no_tools.bind_tools(tools)
```
